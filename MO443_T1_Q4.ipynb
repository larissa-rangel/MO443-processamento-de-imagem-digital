```python
from google.colab import drive

drive.mount('/content/drive')
#Importando todas as bibliotecas necess√°rias
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

```python
dog = io.imread('/content/drive/MyDrive/Testes/little_dog.jpg') #Carregando a imagem da pasta no drive e armazenando em uma matriz np.array
print(dog.shape, dog.dtype)#Imprimir as dimens√µes da matriz que cont√©m os dados da imagem
print(dog.max(),dog.min()) #Imprimir o valor m√°ximo e m√≠nimo da matriz que cont√©m os dados da imagem
plt.imshow(dog)#Exibir os dados da matriz como uma imagem em  RGB
plt.show()#Exibir a figura aberta.
```

```python
#R‚Äô = 0.393R + 0.769G + 0.189B
#G‚Äô = 0.349R + 0.686G + 0.168B
#B‚Äô = 0.272R + 0.534G + 0.131B
new_dog = np.zeros(dog.shape, dtype='uint8') #Inicializa a matriz com zeros
#Cada canal da matriz new_dog vai receber os valores das pondera√ß√µes de R', G' e B"
#A fun√ß√£o np.minimum compara duas matrizes e retorna uma nova matriz contendo os m√≠nimos elementares e foi usada para tratar um overflow que estava ocorrendo
#Por ultimo foi usado o astype para converter os valores para inteiros de 8 bits
new_dog[:,:,0] = np.minimum((0.393*dog[:,:,0]+0.769*dog[:,:,1]+ 0.189*dog[:,:,2]),255).astype(np.uint8)
new_dog[:,:,1] = np.minimum((0.349*dog[:,:,0]+0.686*dog[:,:,1]+ 0.168*dog[:,:,2]),255).astype(np.uint8)
new_dog[:,:,2] = np.minimum((0.272*dog[:,:,0]+0.534*dog[:,:,1]+ 0.131*dog[:,:,2]),255).astype(np.uint8)
print(new_dog.shape, new_dog.dtype)#Imprimir as dimens√µes da matriz que cont√©m os dados da imagem
print(new_dog.max(),new_dog.min())#Imprimir o valor m√°ximo e m√≠nimo da matriz que cont√©m os dados da imagem
plt.imshow(new_dog)#Exibir os dados da matriz como uma imagem em  RGB
plt.show()#Exibir a figura aberta.
```

```python
new_new_dog = np.zeros(dog.shape, dtype="uint8")#Inicializa a matriz com zeros
#Cada canal da matriz new_new_dog vai receber o valor da pondera√ß√£o
#Aqui tamb√©m foi usada a fun√ß√£o np.minimum e o astype
new_new_dog = np.minimum((0.2989*dog[:,:,0] + 0.5870*dog[:,:,1] + 0.1140*dog[:,:,2]),255).astype(np.uint8)
print(new_new_dog.shape, new_new_dog.dtype)
print(new_new_dog.max(),new_new_dog.min())
plt.imshow(new_new_dog,cmap = 'gray') #Exibir os dados da matriz como uma imagem em  cinza se n√£o usar cmap=gray o imshow deixa a imagem esverdeada
plt.show()
```
