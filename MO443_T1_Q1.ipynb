```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
from skimage.util import view_as_blocks
```

```python
baboon = io.imread('/content/drive/MyDrive/Testes/baboon.png') #realizar a leitura da imagem
#Imprimir a imagem e as dimens√µes
print(baboon.shape, baboon.dtype)
plt.imshow(baboon,cmap='gray')
plt.show()
```

```python
#Matriz de blocos com a nova ordem
new_order = np.array(([6,11,13,3],[8,16,1,9],[12,14,2,7],[4,15,10,5]))
#A fun√ß√£o view_as_blocks divide uma matriz M em n, outras matrizes de tamanho (x,y) definidos pelo usu√°rio
#Sendo n= Tamanho de M/Npeda√ßos
```

```python
n = int(np.shape(baboon)[0]/np.shape(new_order)[0])
B = view_as_blocks(baboon, block_shape=(n,n))
#Criar uma nova imagem de tamanho 512 x 512 para guardar o mosaico
print(np.shape(B))
print(B.max(),B.min())
print (n)
```

```python
n
```

```python
baboon_mosaic = (np.zeros(baboon.shape)).astype(np.uint8)
baboon_mosaic = np.hstack((np.concatenate((B[1,1], B[1,3], B[2,3], B[0,3]), axis=0),
                           np.concatenate((B[2,2], B[3,3], B[3,1], B[3,2]), axis=0),
                           np.concatenate((B[3,0], B[0,0], B[0,1], B[2,1]), axis=0),
                           np.concatenate((B[0,2], B[2,0], B[1,2], B[1,0]), axis=0)))

print(np.shape(baboon_mosaic))
```

```python
print(baboon_mosaic.shape, baboon_mosaic.dtype)
print(baboon_mosaic.max(),baboon_mosaic.min())
plt.imshow(baboon_mosaic,cmap='gray')
plt.show()
```
