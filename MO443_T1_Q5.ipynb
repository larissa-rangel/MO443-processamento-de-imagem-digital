1.5 Ajuste de Brilho

```python
from google.colab import drive
drive.mount('/content/drive')
#Importando todas as bibliotecas necessárias
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

```python
baboon = io.imread('/content/drive/MyDrive/Testes/baboon.png')  #Carregando a imagem da pasta no drive e armazenando em uma matriz np.array
print(baboon.shape, baboon.dtype)#Imprimir as dimensões da matriz que contém os dados da imagem
print(baboon.max(),baboon.min())#Imprimir o valor máximo e mínimo da matriz que contém os dados da imagem
plt.imshow(baboon,cmap='gray')#Exibir os dados da matriz como uma imagem em  cinza
plt.show()#Exibir a figura aberta
```

```python
A  = baboon / 255 #convertendo as intensidades dos pixels para o intervalo de [0, 255] para [0, 1]
gama = np.array((1.5,2.5,3.5)) #criando um vetor com os valores gama
fig = plt.figure(figsize=(15,15)) #cria uma figura de tamanho (15,15)
rows = 1 #linhas
columns = 4 #colunas
for i in range(len(gama)): #este laço serve para fazer os cálculos do ajuste de brilho e plotar todas as imagens de uma vez só
  B = ((A ** (1/gama[i])) *255).astype(np.uint8) #percorre o vetor gama e faz a conta A^1/gama para cada coeficiente
  print(B.shape,B.dtype) #apenas para verificar se a dimensão e o tipo de dado estão corretos
  print(B.max(),B.min()) #apenas para verificar se os valores máximo e mínimo estão dentro do intervalo 0 e 255
  fig.add_subplot(rows, columns, i+1) #adciciona o resultado uma posição na grade do subplot
  plt.title(r'$\gamma$ = '+ str(gama[i])) #imprime o título para cada posição do vetor gama
  plt.imshow(B,cmap='gray')# Exibir os dados da matriz em cinza


plt.show()#Exibir a figura aberta
```
