```python
#Importando todas as bibliotecas necessárias
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

```python
img_A = io.imread('/content/drive/MyDrive/Testes/city.png') #Carregando a imagem da pasta no drive e armazenando em uma matriz np.array
print(img_A.shape, img_A.dtype)#Imprimir as dimensões da matriz que contém os dados da imagem
print(img_A.max(),img_A.min())  #Imprimir o valor máximo e mínimo da matriz que contém os dados da imagem
plt.imshow(img_A,cmap='gray') #Exibir os dados da matriz como uma imagem em  cinza
plt.show() #Exibir a figura aberta.
```

```python
#Letra B: Negativo da imagem
img_B  = abs(255-img_A) #Subtrai 255 de todos os elementos de A e faz o módulo do resultado
print(img_B .shape,img_B .dtype)
print(img_B.max(),img_B.min())
```

```python
#Letra C:Converter o intervalo de intensidades para [100, 200]
img_C = (img_A* (100/255)+100).astype(np.uint8) #Realiza a conversão do intervalo de valores de 0 a 255 para 100 a 200
print(img_C.shape,img_C.dtype)
print(img_C.max(),img_C.min())
```

```python
#Letra D:Inverter os valores dos pixels das linhas pares da imagem, ou seja, os valores dos pixels da linha 0
#serao posicionados da direita para esquerda, os valores dos pixels da linha 2 serao posicionados
#da direita para a esquerda e assim por diante
img_D = np.zeros((img_A.shape)) #inicializa uma matriz img_D de mesma dimensão de img_A com zeros
for i in range(len(img_A)):
    if i%2==0:
        img_D[i,:]= img_A[i,::-1]#Este laço percorre cada linha da matriz img_A e testa! Se for par img_D recebe a
    else:                        #linha refletida de img_A, se não img_D recebe a cópia da linha de img_A
        img_D[i,:]= img_A[i,:]

img_D = img_D.astype(np.uint8)
print(img_D.shape,img_D.dtype)
print(img_D.max(),img_D.min())
```

```python
##Letra E: Espelhar as linhas da metade superior da imagem na parte inferior da imagem
img_E = np.vstack((img_A[0:int(len(img_A)/2)],np.flipud(img_A[0:int(len(img_A)/2)]))) #Concatena verticalmente metade da matriz img_A com
print(img_E.shape,img_E.dtype)                                                        #sua outra metade invertida de cima para baixa usando
print(img_E.max(),img_E.min())                                                        #np.flipud
```

```python
#Letra F: Aplicar um espelhamento vertical na imagem levando-se em conta todas as linhas da imagem
img_F = np.flipud(img_A)#Inverte de cima para baixo as linhas da matriz passada como parâmetro
print(img_F .shape,img_F .dtype)
print(img_F.max(),img_F.min())
#print(img_A[511,:]==img_F[0,:]) #verifica se a última linha de A é igual a primeira linha de F
```

```python
fig = plt.figure(figsize=(10,10))#Criando uma nova figura
rows = 2 #linhas
columns = 3 # colunas
fig.add_subplot(rows, columns,1) #Adicionando um eixo de uma linha e 3 colunas para plotar as imagens juntas na mesma figura
plt.imshow(img_A,cmap='gray')
plt.title("Imagem Original ")
fig.add_subplot(rows, columns, 2)
plt.imshow(img_B,cmap='gray')
plt.title("Negativo da imagem")
fig.add_subplot(rows, columns, 3)
plt.imshow(img_A,cmap='gray',vmin=100, vmax=200)#Foi usado o vmin e vmax pois o imshow por default volta os valores de volta para o e 255
plt.title("Imagem transformada")
fig.add_subplot(rows, columns, 4)
plt.imshow(img_D,cmap='gray')
plt.title("Linhas pares invertidas")
fig.add_subplot(rows, columns, 5)
plt.imshow(img_E,cmap='gray')
plt.title("Reflexao de linhas ")
fig.add_subplot(rows, columns, 6)
plt.imshow(img_F,cmap='gray')
plt.title("Espelhamento vertical")
```
