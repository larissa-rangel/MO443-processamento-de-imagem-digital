**1.2 Combinacão de Imagens **
Combinar duas imagens monocromaticas de mesmo tamanho por meio da média ponderada de seus níveis de cinza.

```python
#Importando todas as bibliotecas necessárias
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

```python
A = io.imread('/content/drive/MyDrive/Testes/baboon.png') #Carregando a imagem da pasta no drive e armazenando em uma matriz np.array
print(A.shape, A.dtype)#Imprimir as dimensões da matriz que contém os dados da imagem
plt.imshow(A,cmap='gray') #Exibir os dados da matriz como uma imagem em  cinza
plt.show()#Exibir a figura aberta.

B = io.imread('/content/drive/MyDrive/Testes/butterfly.png')  #Carregando a imagem da pasta no drive e armazenando em uma matriz np.array
print(B.shape, B.dtype)#Imprimir as dimensões da matriz que contém os dados da imagem
plt.imshow(B,cmap='gray')#Exibir os dados da matriz como uma imagem em cinza
plt.show()#Exibir a figura aberta.
```

```python
C = (0.2*A + 0.8*B).astype(np.uint8) #Realizando a combinação das duas matrizes e salvando em tipo np.uint8
print(C.shape, C.dtype)#Imprimir as dimensões da matriz
print(C.min(), C.max())#Imprimir os valores min e máx da matriz
D = (0.5*A + 0.5*B).astype(np.uint8)
print(D.shape, D.dtype)
print(D.min(), D.max())
E = (0.75*A + 0.35*B).astype(np.uint8)
print(E.shape, E.dtype)
print(E.min(), E.max())
```

```python
fig = plt.figure(figsize=(10,10)) #Criando uma nova figura
rows = 1 #linhas
columns = 3 # colunas
fig.add_subplot(rows, columns,1)  #Adicionando um eixo de uma linha e 3 colunas para plotar as imagens juntas na mesma figura
plt.imshow(C,cmap='gray',vmin=C.min(), vmax=C.max())#Exibir os dados da matriz como uma imagem em cinza dentro dos valores vmin e vmax
fig.add_subplot(rows, columns, 2)
plt.imshow(D,cmap='gray',vmin=D.min(), vmax=D.max())
fig.add_subplot(rows, columns, 3)
plt.imshow(E,cmap='gray',vmin=E.min(), vmax=E.max())

```
